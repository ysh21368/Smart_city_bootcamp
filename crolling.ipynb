```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from tqdm import tqdm

col = ['지역', '매장명', '현황', '주소', '전화번호']
hollys_list = []

for i in tqdm(range(1, 53)):
    url = f'https://www.hollys.co.kr/store/korea/korStore2.do?pageNo={i}&sido=&gugun=&store='
    r = requests.get(url)
    soup = BeautifulSoup(r.text, 'html.parser')
    tbody = soup.find('tbody')
    if tbody is None:
        continue
    
    text = tbody.get_text(strip=True)
    temp = [x for x in text.split('\n') if x]
    temp_list = [temp[j::5] for j in range(5)]
    hollys_df = pd.DataFrame(temp_list).T
    hollys_df.columns = col
    hollys_list.append(hollys_df)

final_df = pd.concat(hollys_list, ignore_index=True)
```

## hollys 매장 정보 크롤링

```python
import pandas as pd
from tqdm import tqdm
```

```python
col = columns=['지역','매장명','현황','주소','전화번호']
for i in tqdm(range(1,53)):
    url = f'https://www.hollys.co.kr/store/korea/korStore2.do?pageNo={i}&sido=&gugun=&store='
    r = requests.get(url)
    soup = BeautifulSoup(r.text,'html.parser')
    text = soup.find('tbody').text
    temp =[]
    for j in text.split("\n"):
        if j != "":
            temp.append(j)
    temp_list= []
    for k in range(5):
        temp_list.append(temp[k::5])
    globals()[f'hollys_{i}'] = pd.DataFrame(temp_list, col).T
```

```python
hollys_list = [globals()[f'hollys_{i}'] for i in range(1, 53)]
```

```python
all_hollys = pd.concat(hollys_list,ignore_index=True)
```

```python
page = [47,48,49]
for i in page:
    all_hollys.at[i, '전화번호'] = all_hollys.at[i, '주소']
    all_hollys.at[i, '주소']    =  all_hollys.at[i, '현황']
    all_hollys.at[i, '현황']     = all_hollys.at[i, '매장명']
    all_hollys.at[i, '매장명'] = None
```
